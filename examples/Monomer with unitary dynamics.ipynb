```python
import numpy as np
from qspectra import *
```

```python
bath = DebyeBath(CM_K * 1, None, None)

monomer_ham = VibronicHamiltonian(
    ElectronicHamiltonian([[11500]], dipoles=[[1, 0, 0]], bath=bath),
    n_vibrational_levels=[5], vib_energies=[200], elec_vib_couplings=[[-100]])
monomer = UnitaryModel(monomer_ham, hilbert_subspace='ge', unit_convert=CM_FS)
```

## Excited state dynamics

```python
monomer_ham.n_states('e')
```

```python
%%time 
t, rho = simulate_dynamics(monomer, unit_vec(0, 5 ** 2), 1000, 'ee')
```

```python
plot(t, np.einsum('tii->ti', rho.reshape(-1, 5, 5)).real);
```

## Absorption spectra

```python
%%time
f, X = absorption_spectra(monomer, 10000, correlation_decay_time=1000)
```

```python
plot(f, X)
xlim(11000, 12000)
```
